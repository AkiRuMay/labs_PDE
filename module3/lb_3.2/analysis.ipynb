```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Загрузка данных с пропуском первого столбца
df = pd.read_csv('data.csv').iloc[:, 1:]

def plot_horizontal_bars(values, labels, heading, color_scheme='plasma'):
    """Визуализация горизонтальных бар-чартов с адаптивными размерами"""
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.barplot(
        x=values,
        y=labels,
        hue=labels,
        palette=color_scheme,
        legend=False,
        ax=ax
    )
    ax.set_title(heading, pad=15)
    ax.set_xlabel('Количество ответов', labelpad=10)
    ax.set_ylabel('Варианты выбора', labelpad=10)
    fig.subplots_adjust(left=0.3, right=0.95, top=0.9, bottom=0.1)

def process_composite_answers(response_series, question_text):
    """Обработка вопросов с множественными ответами"""
    split_responses = response_series.dropna().str.split(';')
    flattened_responses = split_responses.explode()
    response_counts = flattened_responses.value_counts()
    
    print(f"\nРезультаты анализа: {question_text}")
    print(response_counts.to_string())
    
    plot_horizontal_bars(
        response_counts.values,
        response_counts.index,
        f"Распределение ответов: {question_text}"
    )
    plt.show()

def handle_rating_questions(response_series, question_text):
    """Анализ вопросов с рейтинговой шкалой"""
    average_score = response_series.mean()
    frequency_distribution = response_series.value_counts().sort_index()
    
    print(f"\nАнализ рейтинга: {question_text}")
    print(f"Средняя оценка: {round(average_score, 2)}")
    print(frequency_distribution.to_string())
    
    fig, ax = plt.subplots(figsize=(8, 5))
    sns.countplot(
        x=response_series,
        hue=response_series,
        palette='rocket',
        legend=False,
        order=sorted(response_series.unique()),
        ax=ax
    )
    ax.set_title(f"Распределение оценок: {question_text}", pad=15)
    ax.set_xlabel('Баллы', labelpad=10)
    ax.set_ylabel('Количество участников', labelpad=10)
    fig.subplots_adjust(left=0.15, right=0.95, top=0.9, bottom=0.15)
    plt.show()

def handle_exclusive_choice(response_series, question_text):
    """Обработка вопросов с единственным выбором"""
    choice_counts = response_series.value_counts()
    
    print(f"\nСтатистика для: {question_text}")
    print(choice_counts.to_string())
    
    plot_horizontal_bars(
        choice_counts.values,
        choice_counts.index,
        f"Распределение ответов: {question_text}",
        'mako'
    )
    plt.show()

def perform_comprehensive_analysis(dataset):
    """Основная функция анализа данных"""
    for column in dataset:
        print(f"\n=== Анализ колонки: {column} ===")
        current_data = dataset[column]
        
        if pd.api.types.is_string_dtype(current_data):
            if current_data.str.contains(';', na=False).any():
                process_composite_answers(current_data, column)
            else:
                handle_exclusive_choice(current_data, column)
        elif np.issubdtype(current_data.dtype, np.number):
            handle_rating_questions(current_data, column)
        else:
            print(f"Неподдерживаемый формат данных в колонке: {column}")

# Запуск анализа
perform_comprehensive_analysis(df)
```
